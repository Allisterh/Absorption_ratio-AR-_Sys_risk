```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.decomposition import PCA
from factor_analyzer import FactorAnalyzer

eig = np.linalg.eig
```

```python
Data = pd.read_excel('Data.xlsx', index_col = 0)
Data
```

```python
# ptc_index_Data
ptc_retrun_Data = Data.pct_change()
ptc_retrun_Data = ptc_retrun_Data.dropna()
```

```python
# Standardized return
ptc_retrun_Data = (Data-Data.mean())/Data.std()
ptc_retrun_Data
```

```python
ptc_retrun_Data.corr()
```

```python
all_Data = ptc_retrun_Data.iloc[:,[0,3,4,5,6,7,8]]
capital_Data = ptc_retrun_Data.iloc[:,[1,3,4,5,6,7,8]]
region_Data = ptc_retrun_Data.iloc[:,[2,3,4,5,6,7,8]]
```

# 전국

## Correlation 구하기

```python
corr_data = all_Data.iloc[0:12].corr().iloc[[0],:]
for i in range(1,34):
    corr_data = corr_data.append(all_Data.iloc[i:12+i].corr().iloc[[0],:])

corr_data.index = all_Data.index[12:46]
corr_data = corr_data.drop(['전국_매매가격지수'], axis=1)
```

## Eigenvalue 구하기

```python
eigen_Data = np.transpose(pd.DataFrame(eig(all_Data.iloc[0:12].corr())[0]))
for i in range(1,34):
    eigen_Data = eigen_Data.append(np.transpose(pd.DataFrame(eig(all_Data.iloc[0+i:12+i].corr())[0])))
eigen_Data.index = all_Data.index[12:46]
eigen_Data
```

## Absorption Raio 구하기

```python
ar_Data = eigen_Data/7
```

## PCA 구하기

```python
# Data에 대해서 PCA 구하기
pca = PCA(n_components=7)
all_pca = pca.fit_transform(all_Data)
pcas_all = pd.DataFrame(pca.explained_variance_ratio_.cumsum()).T
for i in range(1, 34):
    all_pca = pca.fit_transform(all_Data.iloc[i:12+i])
    pcas_all = pcas_all.append(pd.DataFrame(pca.explained_variance_ratio_.cumsum()).T)
pcas_all.index = all_Data.index[12:46]
```

```python
corr_data.to_excel('결과/전국 상관관계 3년.xlsx')
eigen_Data.to_excel('결과/전국 eigenvalue 3년.xlsx')
ar_Data.to_excel('결과/전국 absorption ratio 3년.xlsx')
pcas_all.to_excel('결과/전국 Cumulative Risk Fraction 3년.xlsx')
```
