```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.decomposition import PCA
from factor_analyzer import FactorAnalyzer

eig = np.linalg.eig
```

```python
Data = pd.read_excel('C:/Users/izzac/Desktop/기업별 Sys_risk/데이터_고정이하여신비율.xlsx', index_col = 0)
```

```python
# ptc_index_Data
Data = Data.pct_change()
Data = Data.dropna()
```

```python
# Standardized return
z_Data = (Data-Data.mean())/Data.std()
```

## PCA

```python
#2년

pca = PCA(n_components=8)
all_pca = pca.fit_transform(z_Data)
pcas_all = pd.DataFrame(pca.explained_variance_ratio_.cumsum()).T
for i in range(1, 43):
    all_pca = pca.fit_transform(z_Data.iloc[i:8+i])
    pcas_all = pcas_all.append(pd.DataFrame(pca.explained_variance_ratio_.cumsum()).T)
pcas_all.index = z_Data.index[8:51]
pcas_all.to_excel('C:/Users/izzac/Desktop/기업별 Sys_risk/결과/고정이하여신비율 Cumulative Risk Fraction 2년.xlsx')
```
